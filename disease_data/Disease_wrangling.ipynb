*Cleaned .xls file manually to remove columns/rows not able to be parsed by pandas.*
*Also removed ailments clearly not caused by food like injuries, exposure to mechanical forces, natural disasters, other unintentional injuries, intentional injuries, self-harm, interpersonal violence, collective violence and legal intervention, drowning, falls, road injury, unintentional injuries, injuries, rabies, yellow fever, dengue and other vector diseases, SIDS, 
.*

```python
import pandas as pd
import numpy as np

disease_df = pd.read_excel("cleanedGHE2016_DALYs-2016-country.xlsx", encoding='latin-1')
columns = disease_df.iloc[:,0].tolist()
disease_df.tail()
```

```python

del disease_df['GHE cause']
country_df = disease_df.transpose()
country_df.columns = columns
#country_df.sort_values('')
country_df.head(10)
```

*Check one country*

```python
United_States = country_df.loc[["United States of America"], :]
United_States.dropna()

United_States.sort_values(by='United States of America', axis=1, ascending=False)
```

*Find top N diseases for each country*


First I separated out the selected countries for which we have sufficient recipes, and then sorted them to get the top diseases. The first column is just the population data, which is necessary to include as it allows calculation of percent of population with each disease. *The Carribean is made up of 26 islands and sovereign nations, but not all are listed in the WHO database. For instance Puerto Rico is considered part of teh Carribean but is officially United States territory, so it is not listed alone on the WHO database*

```python
#Caribbean Countries
Caribbean_countries = country_df.loc[["Cuba", "Haiti", "Dominican Republic", "Jamaica", "Trinidad and Tobago", "Bahamas", "Barbados", "Saint Lucia", "Saint Vincent and the Grenadines", "Grenada", "Antigua and Barbuda"]]

#need to sum each column and make it the column for combined dataframe for Carribean
Caribbean_countries.loc['Caribbean'] = pd.Series(Caribbean_countries.sum())
Caribbean_countries

Caribbean = Caribbean_countries.loc[["Caribbean"]].sort_values(by='Caribbean', axis=1, ascending=False)
Caribbean
```
